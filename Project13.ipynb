```python
import pandas as pd
import pymongo
from sqlalchemy import create_engine
```

```python
movie_file = "Resources/netflix_titles.csv"
movies_df = pd.read_csv(movie_file)
movies_df
```

```python
imdb_file = "Resources/IMDb_movies.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
# Create a filtered dataframe from specific columns
netflix_cols = ["show_id", "title", "type", "rating", "release_year", "country", "listed_in"]
netflix_transformed= movies_df[netflix_cols].copy()

netflix_transformed.drop_duplicates("show_id", inplace=True)
netflix_transformed.set_index("show_id", inplace=True)

netflix_transformed
```

```python
imdb_cols = ["imdb_title_id","title","year","avg_vote","votes"]
imdb_transformed= imdb_df[imdb_cols].copy()

imdb_transformed = imdb_transformed.rename(columns={"year": "release_year"})
                                           


imdb_transformed.drop_duplicates("imdb_title_id", inplace=True)
imdb_transformed.set_index("imdb_title_id", inplace=True)

imdb_transformed.head()
```

```python
combined_df = pd.merge(left=netflix_transformed, right=imdb_transformed, on=['title','release_year'], how = 'left')
combined_df.head()
```

```python
merged_movies_only_df = combined_df.loc[combined_df['type'] == "Movie"]
merged_movies_only_df
```

```python
no_null_df =  merged_movies_only_df.dropna(subset = ['avg_vote'])
no_null_df
```

```python
avg_rating_df = no_null_df["avg_vote"].mean()
print(avg_rating_df) 
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

db = client.netflix_db
```

```python
db.movies.insert_many(no_null_df.to_dict('records'))
```
